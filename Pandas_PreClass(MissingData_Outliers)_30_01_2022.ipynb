```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
id_no = ["P001","P002","P003","P004","P005","P006","P007","P008","P009","P010","P011"]
gender=["M","F","M","F","M","F","M","F","M","F","M"]
status = ["FT","PT","-","FT","PT","PT","FT","-","PT","FT",np.nan]
dept = ["DS","FS","AWS","AWS","DS",np.nan,"FS","FS",np.nan,"DS","AWS"]
V1 = np.array([2,3,5,np.nan,7,1,np.nan,10,14,"-",6])
V2 = np.array([8,np.nan,5,8,11,np.nan,np.nan,2,3,7,9])
salary = np.array([np.nan,54,59,120,58,75,np.nan,136,60,125,np.nan])
df = pd.DataFrame({
                   "id" : id_no,
                   "gender": gender,
                   "status": status,
                   "dept": dept,
                   "var1" : V1,
                   "var2" : V2,
                   "salary" : salary
                  })
df
```

# **filling different values with MEAN**

```python
x = df.salary.mean()
x
```

```python
df7 = df.fillna(value= x)
df7
```

```python
df7 = df.fillna(value= df["salary"].mean())  # ortalama değeri ile doldurdum
df7
```

```python
df8 = df.fillna(value= df["salary"].max())  # max değeri ile doldurdum
df8
```

```python
df9 = df.fillna(value= df["salary"].min())  # minimum değeri ile doldurdum
df9
```
