# **BOŞ DEĞERLERİ DOLDURMA**

FİLLNA()
 
 Hepsini dodurmak halı sonuçlar doğurabilir.

```python
df4 = pd.DataFrame({"val1":[2,4,np.nan, 6, np.nan,8,10],
                    "val2": [123,np.nan, 456, np.nan, 789,246,357],
                    "val3": ["France", "Greece", "USA", "Japan", "Sweden","Norway", "Turkey"]})
```

```python
df4
```

```python
df4.fillna(100000)
```

```python
df4["val1"].fillna(1000000)  # sadece val1 deki NAN'ları doldurur
```

```python
df4
```

```python
df4.mean()
```

```python
df4.fillna(df4.mean())                  # val1 deki NAN'ları  mean= 6 ile, 
                                        # val2'deki NAN'ları da mean = 394.2 ile doldurdu (anlamlı bir dodurma yaptık)
```

```python
df4["val3"][0::2]
```

```python
df4["val3"][0::2]= np.nan # uyarı geldi
```

```python
df4["val3"].fillna("Turkey") # boş değerleri Turkey ile doldurduk
```

```python
df4
```

```python
df4.fillna(method="ffill")   #  kendinden bir önceki değer ile doldurdu     " ffill == pad ""
```

```python
df4.fillna(method="pad")      #  kendinden bir önceki değer ile doldurdu     " ffill == pad ""
```

```python
df4.fillna(method="bfill")            #  kendinden bir sonraki değer ile doldurdu     " bfill == backfill ""
```

```python
df4.fillna(method="backfill")           #  kendinden bir sonraki değer ile doldurdu     " bfill == backfill ""
```
