```python
#pip install sqlalchemy
```

```python
from sqlalchemy import create_engine # sql tool arasında bağlantı sağlayacak olan tool
```

```python
temp_db = create_engine("sqlite:///:memory:")   # lokalde(ram de) memory isminde geçici bir DB oluşturdum. Bilgisayarda bulamayız
```

```python
temp_db.table_names()
```

```python
tables = pd.read_html("https://en.wikipedia.org/wiki/World_population")
```

```python
tables
```

```python
tables[0]
```

```python
tables[6]
```

```python
tables[6].to_sql(name="populations", con=temp_db) # populationsa ismi ile temp dB atar
```

```python
temp_db.table_names() # artık görüyor
```

```python
pd.read_sql(sql="populations", con=temp_db)
```

```python
pd.read_sql_query(sql = "SELECT Country, population FROM populations", con=temp_db)
```
