# **AGGREGATION**

```python
import numpy as np
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A','B','C', 'A', 'B', 'C'],
                   'data1': range(6), 
                   'data2': rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df
```

```python
df.groupby('key')['data1'].median()
```

```python
df.groupby('key')['data1'].mean()
```

```python
df.groupby('key')['data1'].max()
```

```python
df.groupby('key')['data1'].min()
```

```python
df.groupby('key')['data1'].agg(['count', 'mean', 'median', 'max', 'min'])
```

```python
df.groupby('key').agg(['count', 'mean', 'median', 'max', 'min'])
```

```python
df.groupby('key')['data1'].agg(['count', 'mean', np.median, 'max', min])
```

```python
df.groupby('key').aggregate({'data1': 'min',
                             'data2': 'max',})
```

# **FILTER()**

```python
df
```

```python
df.groupby('key').std()
```

```python
def filter_func(x):     # function is true for only  B and C, so A is false. 
   return x['data2'].std()>4
```

```python
df.groupby('key').filter(filter_func)
```

```python
display('df', "df.groupby('key').std()" , "df.groupby( 'key').filter(filter_func)")
```

# **TRANSFORM()**

```python
df
```

```python
df.groupby('key').transform(lambda x: x-x.mean())
```

# **APPLY()**

```python
df
```

```python
def norm_by_data2(x):
 # x['data1']/= x['data1'] / x['data2'].sum()
  x['data1']/= x['data2'].sum()
  return x
```

```python
df.groupby('key').apply(norm_by_data2)
```

```python
display('df', "df.groupby('key').apply(norm_by_data2)")
```
